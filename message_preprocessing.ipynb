```python
import pandas as pd
```

## load data

```python
message = pd.read_csv('./UTF8/재난문자.csv')
message.head()
```

```python
# 기간(2020.1~8)까지 자르기
year_2020 = message['연']==2020
year = message[year_2020]

month_14 = (year['월'] < 9)
month = year[month_14]

month
```

## keyword: 코로나
'코로나'라는 키워드가 있는 메세지만 추출

```python
# 내용 column에서 '코로나' 키워드가 있는 데이터 추출
contains_corona = month['내용'].str.contains("코로나")
```

```python
corona_message = month[contains_corona]

corona_message.head()
```

```python
## "코로나" 키워드를 포함한 그룹...
# 각 구, 월끼리 묶어 count 
group_corona = corona_message.groupby(['수신지역','월']).count()

#df정리
group_corona = group_corona.drop(['연','일','시','분'], axis=1)
group_corona.columns = ["count"]

#주 지역구 9개(강남,송파,서초,영등포,중구,마포,강서,구로,용산)추출
key_corona_group = group_corona.loc[['강남구', '송파구', '서초구', '영등포구', '마포구', '강서구', '구로구', '용산구'],:]

key_corona_group
```

## monthly count
구별/월별 메세지 수

```python
## 코로나와 관련 없이 메세지만 카운드.
# 각 구, 월끼리 묶어 count 
group_corona = month.groupby(['수신지역','월']).count()

#df정리
group_corona = group_corona.drop(['연','일','시','분'], axis=1)
group_corona.columns = ["count"]

#주 지역구 9개(강남,송파,서초,영등포,중구,마포,강서,구로,용산)추출
origin_group = group_corona.loc[['강남구', '송파구', '서초구', '영등포구', '중구', '마포구', '강서구', '구로구', '용산구'],:]

origin_group

# group_corona
```

## datetime(실패)
datetime 월별/구별 메세지 수
- datetime 변환 후 resample함수로 묶으려 했으나... 멀티인덱스?실패함

```python
import numpy as np
```

```python
# origin_group을 Datetime 형식으로 바꾼 것
month['Date'] = month['연'].map(str)+'0'+month['월'].map(str)+month['일'].map(str)
month['Date'] = pd.to_datetime(month['Date'], format="%Y%m%d")

d_month = month.drop(['연','월','일','시','분','내용'], axis=1)
d_month['count'] = 1 # count를 위한 column 추가

# 그룹화
df_month = d_month.groupby(['Date','수신지역']).count()

# 월별 카운트
monthly_count = df_month.resample('M',level=0).sum()
```
