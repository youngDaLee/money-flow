```python
import pandas as pd
```

```python
message = pd.read_csv('./UTF8/재난문자.csv')
message.head()
```

```python
# 내용 column에서 '코로나' 키워드가 있는 데이터 추출
contains_corona = message['내용'].str.contains("코로나")
```

```python
corona_message = message[contains_corona]

corona_message.head()
```

```python
# 기간(2020.1~4)까지 자르기
year_2020 = corona_message['연']==2020
year = corona_message[year_2020]

month_14 = (year['월'] < 5)
month = year[month_14]

month
```

```python
# 각 구, 월끼리 묶어 count 
group_corona = corona_message.groupby(['수신지역','월']).count()

#df정리
group_corona = group_corona.drop(['연','일','시','분'], axis=1)
group_corona.columns = ["count"]

#주 지역구 9개(강남,송파,서초,영등포,중구,마포,강서,구로,용산)추출
origin_group = group_corona.loc[['강남구', '송파구', '서초구', '영등포구', '중구', '마포구', '강서구', '구로구', '용산구'],:]

origin_group

# group_corona
```

```python
# 각 구, 월끼리 묶어 count 
group_corona = month.groupby(['수신지역','월']).count()

#df정리
group_corona = group_corona.drop(['연','일','시','분'], axis=1)
group_corona.columns = ["count"]

#주 지역구 9개(강남,송파,서초,영등포,중구,마포,강서,구로,용산)추출
message_group = group_corona.loc[['강남구', '송파구', '서초구', '영등포구',  '마포구', '강서구',  '용산구'],:]

message_group

# group_corona
```

```python
# csv파일로 내보내기
origin_group.to_csv("origin_corona_message_count.csv", encoding='utf-8-sig')
message_group.to_csv("some_corona_message_count.csv", encoding='utf-8-sig')
```
