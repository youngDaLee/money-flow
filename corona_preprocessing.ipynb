```python
import pandas as pd
```

```python
corona = pd.read_csv('./UTF8/서울특별시 코로나19 자치구별 확진자 발생동향.csv')
corona
```

```python
corona['자치구 기준일']=corona['자치구 기준일'].str[:10]
corona = corona.drop(['수집일'],axis=1)
corona = corona.dropna(axis=0)
```

```python
corona
```

```python
#df[df['주소'].str.contains('서울')] ## df[조건식]
is_all = corona.columns.str.contains('전체')
df_corona = corona.loc[:,is_all]
```

```python
df_corona['Date'] = pd.to_datetime(corona['자치구 기준일'], format="%Y.%m.%d")
```

```python
df_corona.head()
```

자치구 9개 '강남구', '송파구', '서초구', '영등포구', '중구', '마포구', '강서구', '구로구', '용산구'
1. 자치구 9개의 일별 코로나 확진자 수
2. 자치구 9개의 월별 코로나 확진자 수 총합

```python
daily_corona = df_corona[['Date', '강남구 전체', '송파구 전체', '서초구 전체', '영등포구 전체', '중구 전체', '마포구 전체', '강서구 전체', '구로구 전체', '용산구 전체']]
daily_corona.columns = ['Date', '강남구', '송파구', '서초구', '영등포구', '중구', '마포구', '강서구', '구로구', '용산구']
```

```python
daily_corona
```

```python
# 2020-8월 데이터까지만 걸러내기
mask = daily_corona['Date']<'2020-09-01'
daily_corona = daily_corona[mask]
```

```python
daily_corona
```

### 월별로 묶어서...

```python
monthly_df = daily_corona.resample('M', on='Date').sum()
```

```python
monthly_df
```

## csv 내보내기

```python
# csv파일로 내보내기
daily_corona.to_csv("corona_month_count.csv", encoding='utf-8-sig') # 월 숫자로만 나눈 것
monthly_df.to_csv("corona_daily_count.csv", encoding='utf-8-sig') # 메세지 카운트
```
