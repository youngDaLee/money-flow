```python
import pandas as pd
```

## load data
데이터 로드해서 각 월별/총합으로 합치기

```python
card_01_1 = pd.read_csv('./UTF8/지역업종별 카드이용정보_202001(1).csv')
card_01_2 = pd.read_csv('./UTF8/지역업종별 카드이용정보_202001(2).csv')
card_02_1 = pd.read_csv('./UTF8/지역업종별_카드이용정보_202002(1).csv')
card_02_2 = pd.read_csv('./UTF8/지역업종별_카드이용정보_202002(2).csv')
card_03_1 = pd.read_csv('./UTF8/지역업종별_카드이용정보_202003(1).csv')
card_03_2 = pd.read_csv('./UTF8/지역업종별_카드이용정보_202003(2).csv')
card_04_1 = pd.read_csv('./UTF8/지역업종별_카드이용정보_202004(1).csv')
card_04_2 = pd.read_csv('./UTF8/지역업종별_카드이용정보_202004(2).csv')
```

```python
card_01 = pd.concat([card_01_1, card_01_2])
card_02 = pd.concat([card_02_1, card_02_2])
card_03 = pd.concat([card_03_1, card_03_2])
card_04 = pd.concat([card_04_1, card_04_2])
```

```python
card = pd.concat([card_01, card_02, card_03, card_04])
```

```python
card.head()
```

## datetime 형식으로 변환

```python
card['기준년월']=pd.to_datetime(card['기준년월'], format="%Y%m")
card.head()
```

## csv 내보내기

```python
card.to_csv("card.csv", encoding='utf-8-sig') # 1~4월 데이터
```
